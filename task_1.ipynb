# Task 1: Load and Explore a Dataset

# Dataset Selection:
For this task, I have chosen the "Iris" dataset, which is a classic dataset often used for machine learning and data analysis. The Iris dataset consists of measurements of four features (sepal length, sepal width, petal length, and petal width) from three different species of iris flowers (setosa, versicolor, and virginica).


# Data Loading:
I will use the pandas library to load the dataset into a DataFrame.

```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
df = pd.read_csv(url, names=column_names)
```

```python
print(df.head())
```

# Dataset Overview:
The Iris dataset contains the following columns:

Sepal Length (in cm)
Sepal Width (in cm)
Petal Length (in cm)
Petal Width (in cm)
Species (the species of iris flower)
Summary Statistics:
Now, let's calculate and display summary statistics for the numerical columns (sepal length, sepal width, petal length, and petal width) of the dataset.

```python
summary_stats = df.describe()

print(summary_stats)
```

# Data Visualization:
To better understand the data, let's create some simple visualizations:

Histograms for each numerical feature:

```python
import matplotlib.pyplot as plt
df.hist()
plt.suptitle("Histograms of Iris Dataset Features")
plt.show()
```

# Scatter plot of sepal length vs. sepal width:

```python
plt.scatter(df["sepal_length"], df["sepal_width"])
plt.title("Scatter Plot: Sepal Length vs. Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

# Bar chart of the count of each species:

```python
species_count = df['class'].value_counts()
species_count.plot(kind='bar', color=['blue', 'green', 'red'])
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Bar Chart of Iris Species')
plt.show()
```

# Documentation:
The Iris dataset is a well-known dataset in the field of machine learning and is often used for classification tasks. It contains 150 samples from three species of Iris flowers (setosa, versicolor, and virginica), with 50 samples from each species. Each sample has four features: sepal length, sepal width, petal length, and petal width, all measured in centimeters.

# Initial Observations:

The dataset is balanced, with an equal number of samples for each species.
The sepal and petal measurements vary in their ranges, with sepal lengths and widths generally having a wider range than petal lengths and widths.
The scatter plot of sepal length vs. sepal width shows some overlap between the species, especially between versicolor and virginica, making it challenging to separate them based solely on these two features.
The bar chart displays the count of each species, showing that there are 50 samples for each species in the dataset.
